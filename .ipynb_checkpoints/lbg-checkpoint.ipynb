```python
from matplotlib import pyplot as plt
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
plt.plot(years, gdp, color='green', marker='o', linestyle='solid') 
plt.title('PIB Nominal') 
plt.ylabel('Bilhões de Dólares') 
plt.show()
```

```python
movies = ['Annie Hall','Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscar = [5,11,3,8,10]
plt.bar(range(len(movies)),  num_oscar)
plt.title('Filmes')
plt.ylabel('Nº de Oscars')
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
from matplotlib import pyplot as plt
from collections import Counter
grades = [80,95,91,87,70,0,85,100,67,73,77,0]
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()],
    histogram.values(),
    10,
    edgecolor=(0,0,0))
plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Notas')
plt.ylabel('N] de Alunos')
plt.title('Distribuição de Notas da 1ª Avaliação')
plt.show()
```

```python
from matplotlib import pyplot as plt
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_erros = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-', label='bias^2')
plt.plot(xs, total_erros, 'b',label='total error')

plt.legend(loc=9)
plt.xlabel('Complexidade do modelo')
plt.xticks([])
plt.title('positori')
plt.show()
```

```python
friends = [70,65,72,63,71,64,60,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends,minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
        xy=(friend_count, minute_count),
        xytext=(5,-5),
        textcoords='offset points')
                 
plt.title('Minutos Diários x Números de Amigos')
plt.xlabel('nº de amigos')
plt.ylabel('minutos dia´rios dedicados ao site')
plt.show()
```
